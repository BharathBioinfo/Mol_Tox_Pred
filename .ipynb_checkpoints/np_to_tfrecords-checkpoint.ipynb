```python
def list_label (path_CSV, file_CSV, path_Dataset):
    csv_file = path_CSV + file_CSV
    
    data_CSV = np.genfromtxt(csv_file, delimiter = ',', usecols = (1))
    data_CSV = data_CSV.tolist()
    
    labels = [1 if i == 1.0 else 0 for i in data_CSV ]
    
    one_hot = np.load(path_Dataset) 
    one_hot = one_hot.item()
    one_hot = one_hot['onehots']
    
    return one_hot, labels
    
    
```
